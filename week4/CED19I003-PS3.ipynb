### Question-3

For a sample space of 15 people, a statistician wanted to know the consumption of water and other beverages.
He collected their average consumption of water and beverages for 30 days (in litres). Help him to visualize the
data using density plot, rug plot and identify the mean, median, mode and skewness of the data from the plot.




```python
water = np.array([3.2, 3.5, 3.6, 2.5, 2.8, 5.9, 2.9, 3.9, 4.9, 6.9, 7.9, 8.0, 3.3, 6.6, 4.4])
beverages = np.array([2.2, 2.5, 2.6, 1.5, 3.8, 1.9, 0.9, 3.9, 4.9, 6.9, 0.1, 8.0, 0.3, 2.6, 1.4])

plt.figure(figsize = (20, 5))

plt.subplot(1, 2, 1)
fig = sns.kdeplot(water, fill=True, color='orange')
fig = sns.rugplot(water, color = 'orange')
fig = sns.kdeplot(beverages, fill=True, color='blue')
fig = sns.rugplot(beverages, color = 'blue')

plt.subplot(1, 2, 2)
fig = sns.kdeplot(water, fill=True, color='orange')
fig = sns.kdeplot(beverages, fill=True, color='blue')
plt.axvline(x = np.mean(water), color = 'orange')
plt.axvline(x = np.mean(beverages), color = 'blue')

plt.plot()
```

### Question-4

A car company wants to predict how much fuel different cars will use based on their masses. They took a
sample of cars, drove each car 100km, and measured how much fuel was used in each case (in litres). Visualize
the data using scatterplot and also find co-relation between the 2 variables (eg. Positive//Negative, Linear/
Non-linear co-relation) The data is summarized in the table below.


```python
Mass = np.array([0.45, 0.91, 1.36, 1.81, 2.27])
Fuel = np.array([3.6, 6.7, 9.8, 11.2, 14.7])

plt.scatter(Mass, Fuel)
plt.plot(np.unique(Mass), np.poly1d(np.polyfit(Mass, Fuel, 1)) (np.unique(Mass)), color='yellow')

plt.xlabel(" Mass (metric ton) ")
plt.ylabel(" Fuel (L) ")

```

### Question-6

Generate random numbers from the following distribution and visualize the data using violin plot.

(i) Standard-Normal distribution.

(ii) Log-Normal distribution.

```python
plt.figure(figsize = (15, 5))

snorm = np.random.normal(size=10)
print("Standard-Normal : ", snorm)
plt.subplot(2, 1, 1)
fig = sns.violinplot(snorm, orient='h', color='orange')

lnorm = np.random.lognormal(size=10)
print("Log-Normal : ", lnorm)
plt.subplot(2, 1, 2)
fig = sns.violinplot(lnorm, orient='h', color='green')
```

### Question-7

An Advertisement agency develops new ads for various clients (like Jewellery shops, Textile shops). The
Agency wants to assess their performance, for which they want to know the number of ads they developed in
each quarter for different shop category. Help them to visualize data using radar/spider charts.

```python
indexnames = ["Textile", "Jewellery", "Cleaning Essentials", "Cosmetics"]
q1 = np.array([10, 5, 15, 14])
q2 = np.array([6, 5, 20, 10])
q3 = np.array([8, 2, 16, 21])
q4 = np.array([13, 4, 15, 11])

data = pd.DataFrame({"Quarter 1" : q1, "Quarter 2" : q2, "Quarter 3" : q3, "Quarter 4" : q4}, index = indexnames)
data[''] = data['Quarter 1']
print(data)


angle = np.linspace(0, 2*np.pi, len(data.columns))

plt.figure(figsize=(8, 8))
plt.subplot(polar=True)

plt.plot(angle, data.iloc[0], label="Textile", linewidth=1, color='blue')
plt.fill(angle, data.iloc[0], 'blue', alpha=0.25)

plt.plot(angle, data.iloc[1], label="Jewellery", linewidth=1, color='green')
plt.fill(angle, data.iloc[1], 'green', alpha=0.25)

plt.plot(angle, data.iloc[2], label="Cleaning Essentials", linewidth=1, color='orange')
plt.fill(angle, data.iloc[2], 'orange', alpha=0.25)

plt.plot(angle, data.iloc[3], label="Cosmetics", linewidth=1, color='red')
plt.fill(angle, data.iloc[3], 'red', alpha=0.25)

labels = plt.thetagrids(np.degrees(angle), data.columns)
plt.legend()
```
